## 自然数の二進表現: Binary-Representation-of-Natural-Numbers
- 自然数 $n$ を二進数で表現
    - $n$ を 2 で割った余りを記録
    - $n$ を 2 で割った商を新しい $n$ とする
    - $n$ が 0 になるまで繰り返す
- 余りを逆順に並べると二進表現になる
----
- Convert a natural number $n$ to its binary representation:
    - Record the remainder when $n$ is divided by 2.
    - Update $n$ to be the quotient of $n$ divided by 2.
    - Repeat until $n$ becomes 0.
- The binary representation is obtained by reversing the order of the recorded remainders.

```python
def convert_to_binary(n:int) -> str:
    """Convert a natural number n to its binary representation as a string.
    """
    if n == 0:
        return "0"
    bits = ""
    while n > 0:
        b = n % 2
        bits = str(b) + bits
        n //= 2
    return bits
```

## 最大公約数: Greatest Common Divisor
- 2つの整数$n$と$m$の最大公約数$r$を求める
- 最初に簡単だが、非効率な方法を示し、次にユークリッドの互除法を説明する
----
- Find the greatest common divisor $r$ of two integers $n$ and $m$
- First, we show a simple but inefficient method, and then explain the Euclidean algorithm
----

### 単純な方法: Simple Method
- $n>m$とする
- for $r$ in $m, m-1, m-2, \ldots, 1$ do 
    - if $n \% r == 0$ かつ $m \% r == 0$
        - $r$が最大公約数
----
- Assume $n > m$
- for $r$ in $m, m-1, m-2, \ldots, 1$ do 
    - if $n \% r == 0$ and $m \% r == 0$
        - $r$ is the greatest common divisor


```python
def simple_gcd(n: int, m: int) -> int:
    """
    2つの整数nとmの最大公約数を求める（非効率な方法）
    
    Find the greatest common divisor of two integers n and m (inefficient method)
    """
    min_value: int = min(n, m)
    # min_valueから1まで降順にループ
    # Loop from min_value down to 1
    for i in range(min_value, 0, -1):
        if n % i == 0 and m % i == 0: # i が公約数: i is a common divisor
            return i
    return 1

# 使用例 / Example usage
n = 48
m = 18
gcd: int = simple_gcd(n, m)
print(f"gcd of {n} and {m} is {gcd}.")
```

この方法では、最悪の場合で$m$回の割り算が必要

In the worst case, this method requires $m$ divisions

----

### ユークリッドの互除法: Euclidean Algorithm
- $n>m$とする
- $r = m$
- while $r \ne 0$ do
    - $r = n \% m$
    - $n  =m$
    - $m = r$
- $m$が最大公約数
---
- Assume $n > m$
- $r = m$
- while $r \ne 0$ do
    - $r = n \% m$
    - $n = m$
    - $m = r$
- $m$ is the greatest common divisor


```python
def euclidean_gcd(n: int, m: int) -> int:
    """
    2つの整数nとmの最大公約数を求める（ユークリッドの互除法）
    
    Find the greatest common divisor of two integers n and m (Euclidean algorithm)
    """
    while m != 0:
        r: int = n % m
        # n = m
        # m = r
        n, m = m, r
    return n

# 使用例 / Example usage
n = 48
m = 18
gcd: int = euclidean_gcd(n, m)
print(f"gcd of {n} and {m} is {gcd}.")
```

### 再帰的ユークリッドの互除法: Recursive Euclidean Algorithm
- ユークリッドの互除法は再帰的に定義できる: The Euclidean algorithm can be defined recursively
$$\begin{align*}
\text{gcd}(n, 0) &= n\\
    \text{gcd}(n, m) &= \text{gcd}(m, r) \quad (r=n \mod m)
\end{align*}$$

```python
def euclidean_gcd_recursive(n: int, m: int) -> int:
    """
    2つの整数nとmの最大公約数を求める（再帰的ユークリッドの互除法）
    
    Find the greatest common divisor of two integers n and m (recursive Euclidean algorithm)
    """
    if m == 0:
        return n
    else:
        r: int = n % m
        return euclidean_gcd_recursive(m, r)

# 使用例 / Example usage
n = 48
m = 18
gcd: int = euclidean_gcd_recursive(n, m)
print(f"gcd of {n} and {m} is {gcd}.")
```

$n=48$, $m=18$の場合は以下のように計算が進む。
$$
(48, 18) \to (18, 12) \to (12, 6) \to (6, 0)
$$
最大公約数6であり、各回の余りも全て6の倍数である。一般に、$u$を$v$で割った余り$r$は、$r\lt u/2$、つまり各回で調べる数が半分以下になる。最悪の場合は、$n$と$m$が互いに素であり、最大公約数が$1$の場合である。つまり、$\log_2 m$回の割り算で最大公約数が求まる。このようなとき、計算量は$O(\log m)$であるという。

----

For $n=48$ and $m=18$, the calculation proceeds as follows:
$$(48, 18) \to (18, 12) \to (12, 6) \to (6, 0)$$
The greatest common divisor is 6, and all remainders in each step are multiples of 6. In general, the remainder $r$ when $u$ is divided by $v$ satisfies $r < u/2$, meaning that the numbers being considered are halved or less in each step. The worst case occurs when $n$ and $m$ are coprime, i.e., when the greatest common divisor is $1$. In this case, the greatest common divisor can be found in $\log_2 m$ divisions. Thus, the time complexity in such cases is $O(\log m)$.

## 素数: Prime Numbers
- 素数は、1とその数自身以外に約数を持たない自然数
- 2以上の自然数$n$が素数であることを確認するには、$2$から$\sqrt{n}$までの整数で割り切れないことを確かめればよい
----
- A prime number is a natural number greater than 1 that has no divisors other than 1 and itself
- To verify that a natural number $n$ greater than or equal to 2 is prime, it suffices to check that it is not divisible by any integer from $

```python
def is_prime(n: int) -> tuple[bool, str]:
    """
    自然数nが素数であるかどうかを判定する
    
    Determine whether a natural number n is prime

    args:
    -----
    n: int
        判定する自然数 / The natural number to be checked

    returns:
    --------
    tuple[bool, str]
        素数であればTrueとメッセージ、素数でなければFalseとメッセージ / True and message if prime, False and message otherwise
    """
    message = f"{n} is "
    match n:
        case n if n != int(n):
            message += "not an integer."
            return False, message
        case n if n < 2:
            message += "is less than 2."
            return False, message
        case 2:
            return True, message + "a prime number."
        case n if n % 2 == 0:
            return False, message + "divisible by 2."
        case _:
            pass
    
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False,message + f"divisible by {i}."
    return True, message + "a prime number."

n = 41
result, msg = is_prime(n)
print(msg)
```

## エラトステネスの篩: Sieve of Eratosthenes
- ある自然数$N$以下の全ての素数を求める
- 2から$N$までの整数のリストを作成する
- リストの先頭の数$p$を素数として記録し、$p$の倍数をリストから削除する
- リストに数が残っている限り、上記の操作を繰り返す
---
- Find all prime numbers less than or equal to a natural number $N$
- Create a list of integers from 2 to $N$
- Record the first number $p$ in the list as a prime number, and remove all multiples of $p$ from the list
- Repeat the above steps as long as there are numbers remaining in the list

```python
def eratosthenes_sieve(N: int) -> list[int]:
    """
    ある自然数N以下の全ての素数を求める（エラトステネスの篩）
    
    Find all prime numbers less than or equal to a natural number N (Sieve of Eratosthenes)
    """
    numbers: list[int] = list(range(2, N + 1))
    primes: list[int] = []
    
    while numbers:
        p: int = numbers[0]
        primes.append(p)
        numbers = [x for x in numbers if x % p != 0]
    
    return primes

N = 50
primes: list[int] = eratosthenes_sieve(N)
print(f"Primes less than or equal to {N}: {primes}")
```
