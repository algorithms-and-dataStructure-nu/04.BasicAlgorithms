### 演習問題: Exercise
整数のリスト、例えば`[3, 1, 4, 1, 5, 9]`が与えられたとき、最小値のインデックスを返す関数`find_min_index(lst)`を実装しなさい。
最小値が複数ある場合には、最初に出現するインデクスを返しなさい。
複数の例題で動作を確認しなさい。

Given a list of integers, for example `[3, 1, 4, 1, 5, 9]`, implement a function `find_min_index(lst)` that returns the index of the minimum value in the list.
If there are multiple occurrences of the minimum value, return the index of the first occurrence.
Test your function with multiple examples to ensure it works correctly.

### 演習問題: Exercise
アルファベット小文字は、`a`は、コードポイント`97`、`b`はコードポイント`98`、...、`z`はコードポイント`122`で表され、例えば`char`関数を使って、コードポイントから文字を得ることができる。例えば、`char(97)`は`'a'`を返す。
また、`random.randint(a, b)`は、整数`a`から`b`までのランダムな整数を返す。
これらを利用して、ランダムなアルファベット小文字からなる長さ`n`の文字列を生成する関数`generate_random_string(n)`を実装し、動作を確認しなさい。

Given that lowercase letters of the alphabet are represented by code points where `a` is `97`, `b` is `98`, ..., and `z` is `122`, and you can use the `char` function to get a character from its code point (for example, `char(97)` returns `'a'`), and the function `random.randint(a, b)` returns a random integer between `a` and `b`, implement a function `generate_random_string(n)` that generates a random string of length `n` consisting of lowercase letters. Test your function to ensure it works correctly.

### 演習問題: Exercise
与えられた文字列`s`が回文であるかどうかを判定する関数`is_palindrome(s)`を実装しなさい。回文とは、前から読んでも後ろから読んでも同じ文字列のことを指す。例えば、`"radar"`や`"level"`は回文であるが、`"hello"`は回文ではない。

Implement a function `is_palindrome(s)` that determines whether a given string `s` is a palindrome. A palindrome is a string that reads the same forward and backward. For example, `"radar"` and `"level"` are palindromes, while `"hello"` is not.
