自然数$n$に対して、その素因数分解を求める関数`prime_factorization(n)`を以下に示す。
この関数は、$n$の素因数分解の結果を、(素因数, 指数)の組のリストとして返す。例えば、$360=2^3 \times 3^2 \times 5^1$の素因数分解は`[(2, 3), (3, 2), (5, 1)]`と表される。

この関数の戻り値が正しいことを確認する関数`check_prime_factorization(n, factors)`を作成し、、動作を確認しなさい。


```python
def prime_factorization(n:int) -> list[tuple[int, int]]:
    """
    nの素因数分解を行い、(素因数, 指数)の組のリストを返す。
    """
    factors: list[tuple[int, int]] = []
    n_dum = n
    count = 0
    while n % 2 == 0:
        n //= 2
        count += 1
    if count > 0:
        factors.append((2, count))
    
    f = 3
    while f*f <= n_dum:
        count = 0
        while n % f == 0:
            n //= f
            count += 1
        if count > 0:
            factors.append((f, count))
        f += 2
    if n > 1:
        factors.append((n, 1))
       
    return factors
```
